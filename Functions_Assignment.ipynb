7. What are the advantages of using generators over regular functions?
  - Generators are memory-efficient, especially for large datasets, because they generate values one by one instead of creating the entire sequence in memory. They can also represent infinite sequences and are useful in data processing pipelines.
Example

A generator is more memory-efficient for large sequences
def large_list_func(n):
    return [i for i in range(n)]

def large_generator_func(n):
    for i in range(n):
        yield i

import sys

   Regular function creates a full list in memory
my_list = large_list_func(100000)
print(f"Size of list: {sys.getsizeof(my_list)} bytes")

   Generator generates values on demand
my_generator = large_generator_func(100000)
print(f"Size of generator: {sys.getsizeof(my_generator)} bytes")

8. What is a lambda function in Python and when is it typically used?
  -A lambda function is a small, anonymous function defined with the lambda keyword. It can only contain a single expression and is often used for short, simple operations or as an argument to higher-order functions like map(), filter(), and sorted().
Example

   Define a lambda function to add two numbers

print(f"Lambda result: {add(5, 3)}")

   Using lambda with `sorted()` to sort a list of tuples by the second element
pairs = [(1, 'b'), (3, 'a'), (2, 'c')]
sorted_pairs = sorted(pairs, key=lambda x: x[1])
print(f"Sorted pairs: {sorted_pairs}")

10. What is the difference between map(), reduce(), and filter() functions in Python?
  -	map() transforms each item in an iterable by applying a function, returning a new iterable of the same size.
  -	filter() constructs a new iterable containing only the items from the original for which a given function returns True.
  -reduce() cumulatively applies a function to the items of an iterable to reduce it to a single value. It requires importing from the functools module.
Example

from functools import reduce

numbers = [1, 2, 3, 4, 5]

   map(): Doubles each number
doubled = list(map(lambda x: x * 2, numbers))
print(f"map() result: {doubled}")

   filter(): Keeps only even numbers
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(f"filter() result: {evens}")

   reduce(): Calculates the sum
total = reduce(lambda x, y: x + y, numbers)
print(f"reduce() result: {total}")



```python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even_numbers(numbers):
    sum_even = 0
    for num in numbers:
        if num % 2 == 0:
            sum_even += num
    return sum_even
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
import math
def is_prime(number):

    if number <= 1:
        return False


    for i in range(2, int(math.sqrt(number)) + 1):
        if number % i == 0:
            return False  # Found a divisor, so it's not prime
    return True  # No divisors found, so it's prime

def check_primes_in_range(start, end):

    print(f"Checking prime numbers from {start} to {end}:")
    for num in range(start, end + 1):
        if is_prime(num):
            print(f"{num} is prime.")
        else:
            print(f"{num} is not prime.")


check_primes_in_range(1, 200)
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, num_terms):
        if num_terms < 0:
            raise ValueError("Number of terms cannot be negative")
        self.num_terms = num_terms
        self.a, self.b = 0, 1
        self.count = 0
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent

def power_of_two_generator(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_lines(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
```

```python
#10.  Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(string):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, string))
```
